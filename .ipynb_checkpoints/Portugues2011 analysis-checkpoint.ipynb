```python
def back_color():
    plt.axvspan(-0.5, 2.5, color=(0.9,)*3)
    plt.axvspan(8.5, 11.5, color=(0.9,)*3)
    plt.axvspan(5.5, 8.5, color=(0.7,)*3)
    plt.axvspan(11.5, 14.5, color=(0.7,)*3)
    
def cum_stat(key):
    stat = np.array([np.nanmean(exp.exp_stats[key].as_matrix()[1:].reshape(6, 18), 0) for exp in exps if not exp.exclude]).T
    stat = stat/np.nanmean(stat[:3,:], 0)
    return stat
```

```python
stat.shape
```

```python
plt.figure(figsize = (8,8))
plt.subplot(2,2,1)
k = 'mean_bout_len'
stat = cum_stat(k)
plt.errorbar(np.arange(stat.shape[0]), np.nanmean(stat, 1), np.nanstd(stat, 1)/np.sqrt(stat.shape[1]-1))
plt.plot(stat, linewidth = 0.1, c='k')
back_color()
plt.ylim(0.7,1.4)

plt.subplot(2,2,2)
k = 'interbout_t'
stat = cum_stat(k)
plt.errorbar(np.arange(stat.shape[0]), np.nanmean(stat, 1), np.nanstd(stat, 1)/np.sqrt(stat.shape[1]-1))
plt.plot(stat, linewidth = 0.1, c='k')
back_color()
plt.ylim(0.5,1.5)

plt.subplot(2,2,3)
k = 'n_bouts'
stat = cum_stat(k)
plt.errorbar(np.arange(stat.shape[0]), np.nanmean(stat, 1), np.nanstd(stat, 1)/np.sqrt(stat.shape[1]-1))
plt.plot(stat, linewidth = 0.1, c='k')
back_color()
plt.ylim(0.5,1.5)

plt.subplot(2,2,4)
k = 'latency_to_first'
stat = cum_stat(k)
plt.errorbar(np.arange(stat.shape[0]), np.nanmean(stat, 1), np.nanstd(stat, 1)/np.sqrt(stat.shape[1]-1))
plt.plot(stat, linewidth = 0.1, c='k')
back_color()
plt.ylim(0.7,1.4)

```

```python
final_fig = plt.figure()
k = 'interbout_t'
stat = cum_stat(k)
plt.errorbar(np.arange(stat.shape[0]), np.nanmean(stat, 1), np.nanstd(stat, 1)/np.sqrt(stat.shape[1]-1))
plt.plot(stat, linewidth = 0.1, c='k')
back_color()
plt.ylim(0.5,1.5)
plt.xlim((-0.5, 17.5))
plt.xticks(np.arange(1, 17, 3), np.arange(2, 18, 3))
plt.ylabel('relative duration')
plt.xlabel('trial number')
```
